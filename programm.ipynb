```python
import re 
def mask_cc_digits(cc_string,digits_to_keep=4,mask_char="*"):
    cc_string_total=sum(map(str.isdigit,cc_string))
    
    if digits_to_keep >= cc_string_total:
        print("add more numbers to the credit card")
        
    digits_to_mask=cc_string_total - digits_to_keep
    masked_cc_string=re.sub("\d",mask_char,cc_string,digits_to_mask)
    
    return masked_cc_string
```

```python
class VEHICLE:
    def __init__(self,name):
        self.name=name
        
```

```python
class MODAL():
    def __init__(self,modal):
        self.modal=modal
```

```python
class DETAILS(VEHICLE,MODAL):
    def __init__(self,name,modal,seating):
        VEHICLE.__init__(self,name)
        MODAL.__init__(self,modal)
        self.seating=seating
        
        
    def display(self):
        print(f"name of the vehicle : {self.name}, modal : {self.modal}, seating capacity : {self.seating}")
```

```python
car=DETAILS("AUDI","AUTOMATIC",4)
```

```python
car.display()
```

```python
class VEHICLE:
    def __init__(self,name):
        self.name=name
        
```

```python
class CAR(VEHICLE):
    def __init__(self,name,modal,price):
        super().__init__(name)
        self.modal=modal
        self.price=price
        
    def display(self):
        print(f" name of vehicle : {self.name}, modal of vehicle : { self.modal}, price of vehicle : {self.price}")
        
```

```python
class BUS(VEHICLE):
    def __init__(self,name,modal,price):
        super().__init__(name)
        self.modal=modal
        self.price=price
        
    def display(self):
        print(f" name of vehicle : {self.name}, modal of vehicle : { self.modal}, price of vehicle : {self.price}")
        
    
```

```python
class VAN(VEHICLE):
    def __init__(self,name,modal,price):
        super().__init__(name)
        self.modal=modal
        self.price=price
        
    def display(self):
        print(f" name of vehicle : {self.name}, modal of vehicle : { self.modal}, price of vehicle : {self.price}")
        
```
